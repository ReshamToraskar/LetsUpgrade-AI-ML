# Attrition Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset = pd.read_csv("general_data.csv")
dataset.head()
```

```python
dataset.dropna(inplace=True)
dataset
```

```python
dataset["Attrition"].replace(to_replace=("No","Yes"),value=(0,1),inplace=True)
dataset.head(5)
```

```python
dataset["BusinessTravel"].replace(to_replace=("Travel_Rarely","Travel_Frequently","Non-Travel"),value=(0,1,2),inplace=True)
dataset.head(5)
```

```python
dataset["Gender"].replace(to_replace=("Female","Male"),value=(0,1),inplace=True)
dataset.head(5)
```

```python
dataset["MaritalStatus"].replace(to_replace=("Married","Single","Divorced"),value=(0,1,2),inplace=True)
dataset.head(5)
```

```python
dataset.columns
```

```python
Y = dataset.Attrition
X = dataset[['Age','DistanceFromHome','BusinessTravel','MonthlyIncome','Gender','MaritalStatus','PercentSalaryHike','TotalWorkingYears','YearsAtCompany']]
```

```python
import statsmodels.api as sm
X1 = sm.add_constant(X)
Logistic = sm.Logit(Y,X1)
result  = Logistic.fit()
```

```python
result.summary()
```

From above Logit Regression result, the variables namely Age,MonthlyIncome, PercentSalaryHike, TotalWorkingYears and YearsAtCompany have p values less than 0.05, thus all these variables are significantly important for the attrition. Out of these Age, TotalWorkingYears and YearsAtCompany most significant.
