**1. MannnWhitney test**

**Attrition and DistanceFromHome**

```python
from sklearn import preprocessing
from scipy.stats import pearsonr
dataset["Attrition"].replace(to_replace=("No","Yes"),value=(0,1),inplace=True)
dataset.head(5)
```

```python
a1=dataset[dataset["Attrition"]==0]
a1.head(5)
```

```python
a2=dataset[dataset["Attrition"]==1]
a2.head(5)
```

```python
from scipy.stats import mannwhitneyu
stats,p=mannwhitneyu(a1.DistanceFromHome,a2.DistanceFromHome)
print(stats,p)
```

As the P value of 0.4629185205822659 is > 0.05, the H0 is accepted and Ha is rejected.
H0: There is no significant differences in the Distance From Home between attrition (Y) and attirition (N)
Ha: There is significant differences in the Distance From Home between attrition (Y) and attirition (N)
