# Applying KNN algorithm on Titanic Dataset

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
titanic = pd.read_csv('train.csv')
titanic.columns
```

```python
le = preprocessing.LabelEncoder()
le.fit(titanic['Sex'])
titanic['Sex']= le.transform(titanic['Sex'])
le.fit(titanic['Embarked'])
titanic['Embarked']= le.transform(titanic['Embarked'])
le.fit(titanic['Age'])
titanic['Age']= le.transform(titanic['Age'])
le.fit(titanic['Fare'])
titanic['Fare']= le.transform(titanic['Fare'])
titanic.head()
```

# K values for Pclass as dependent variable

```python
from sklearn import neighbors
```

```python
y = titanic['Pclass']
X = titanic.drop(['Pclass','PassengerId','Name','Ticket','Cabin'], axis = 1)
X_train,X_test,y_train,y_test = train_test_split(X, y , test_size=0.3, random_state=0)
def createList(r1, r2): 
    return list(range(r1, r2+1)) 
for j in createList(1,267):
    knn = neighbors.KNeighborsClassifier(n_neighbors = j)
    a = knn.fit(X_train,y_train).score(X_test,y_test)
    y_pred = knn.predict(X_test)
    b = confusion_matrix(y_test, y_pred)
    print("accuracy score for Pclass when k is", j, "is \n", a)
    print("Confusion Matrix for Pclass when k is", j, "is \n", b)
```

###### Accuracy is highest for k = 1 and that is 88% and it's the least 66% for k =198 to k= 229
