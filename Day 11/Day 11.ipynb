**Day 11 Assignment: to check the correlation between Attrition and other variables**

```python
import pandas as pd
dataset = pd.read_csv("general_data.csv")
dataset.columns
dataset.head(10)
```

```python
from sklearn import preprocessing
from scipy.stats import pearsonr
dataset["Attrition"].replace(to_replace=("No","Yes"),value=(0,1),inplace=True)
```

```python
dataset.head(5)
```

**Correlation between Attrition and Distance from home**
H0 = There is no significant correlation between Attrition and DistanceFromHome
Ha = There is significant correlation between Attrition and DistanceFromHome

```python
stats,p = pearsonr(dataset.Attrition,dataset.DistanceFromHome)
print(stats,p)
```

As the p-value is greater than 0.05, we'll accept the null hypothesis.
Also from the value of r, there seems weak negative correlation between Attrition and DitanceFromHome.

```python
dataset["Gender"].replace(to_replace=("Female","Male"),value=(0,1),inplace=True)
dataset.head(5)
```
